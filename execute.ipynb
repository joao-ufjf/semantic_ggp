```python
import asyncio
import asyncio.subprocess
import datetime

start = datetime.datetime.now()

sem = asyncio.Semaphore(15)

commands = None
with open('commands.txt', 'r') as file:
    commands = file.readlines()

executed = None
with open('executed.txt', 'r') as file:
    executed = file.readlines()

async def do_job(c, total, command):
    async with sem:  # Don't run more than 10 simultaneous jobs below
        print(c/total, command)
        proc = await asyncio.create_subprocess_shell(command, stdout=asyncio.subprocess.PIPE)
        output = await proc.stdout.read()

        proc = await asyncio.create_subprocess_shell('''echo "{}" >> executed.txt'''.format(command[:-1]), stdout=asyncio.subprocess.PIPE)
        output = await proc.stdout.read()
    
tasks = []
total = len(commands)
for c, command in enumerate(commands[:]):
    if command not in executed:
        task = asyncio.create_task(do_job(c, total, command))
        tasks.append(task)

await asyncio.gather(*tasks)

print(datetime.datetime.now() - start)
```
